```python
from google.colab import files


uploaded = files.upload()

import pandas as pd

df = pd.read_csv("sales.csv")


df.head()
print(df.info())
print(df.describe())


sales_by_product = df.groupby("Product")["Sales"].sum()
print(sales_by_product)


sales_by_region = df.groupby("Region")["Sales"].sum()
print(sales_by_region)


sales_by_date = df.groupby("Date")["Sales"].sum()
print(sales_by_date)

import matplotlib.pyplot as plt


sales_by_product.plot(kind="bar", title="Total Sales by Product")
plt.ylabel("Sales")
plt.show()


sales_by_region.plot(kind="pie", autopct="%1.1f%%", title="Sales by Region")
plt.ylabel("")
plt.show()


sales_by_date.plot(kind="line", figsize=(10,5), title="Sales Trend Over Time")
plt.ylabel("Sales")
plt.show()

best_product = sales_by_product.idxmax()
print("Best selling product:", best_product)

best_region = sales_by_region.idxmax()
print("Top region:", best_region)
```
